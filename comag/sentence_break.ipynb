```python
import pickle
import nltk
```

```python
topic = 'ugip'
```

```python
data = pickle.load(open("pickles/"+topic+".data.t1", "r"))
```

```python
data
```

```python
sentence_pairs = []
```

```python
for x in data:
    sentences = nltk.sent_tokenize(x['comment'])
    sentence_pairs.append((sentences, x['reason'], x['id']))
```

```python
sentence_pairs
```

```python
pickle.dump(sentence_pairs, open("pickles/"+topic+".sentences.t1", "w"))
```

# Build STS features

```python
from requests import get
```

```python
sss_url = "http://swoogle.umbc.edu/StsService/GetStsSim"
```

```python
def sss(s1, s2):
    try:
        response = get(sss_url, params={'operation':'api','phrase1':s1,'phrase2':s2})
        return float(response.text.strip())
    except:
        print 'Error in getting similarity for %s: %s' % ((s1,s2), response)
        return 'a'
```

```python
failed = []
sts = []
for i, x in enumerate(sentence_pairs):
    sts_x = []
    for j, s in enumerate(x[0]):
        val = sss(s, x[1])
        if val == 'a':
            failed.append((i,j))
        else:
            sts_x.append((i,j,val))
    sts.append(sts_x)
```

```python
pickle.dump(sts, open("pickles/"+topic+".sts", "w"))
```
